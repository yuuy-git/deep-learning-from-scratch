## NumPy

```python
import numpy as np
```

NumPyの配列の計算は要素ごと（element-wise）に行われる。

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
print('x + y = ', x+y)
print('x * y = ', x*y)
```

element-wiseな計算だけでなくスカラ値との計算も可能。  
この機能をブロードキャストという。  
下の例では10というスカラ値が1×3の行列 [10, 10, 10] として扱われている。

```python
x = np.array([1, 2, 3])
print('x * 10 = ', x*10)
```

以下の例からブロードキャストでの計算は、行列とスカラー値間での計算、  
または次元数a, bに対して `a = n*b` の関係が成立する行列間にのみ計算可能であることがわかる。

```python
x = np.array([[1, 2], [3, 4]])
y = np.array([10, 20])
z = np.array([10, 20, 30])
print('x * y = ', x*y)
print('x * z = ', x*z)
```

for文では各行ごとにアクセスされるらしい。

```python
X = np.array([[51, 55], [14, 19], [0, 4]])
for row in X:
    print(row)
```

各要素へのアクセス方法として、以下のように条件を指定することもできる。

```python
Y = X[X>15]
print(Y)

Z = X[X%2 == 0]
print(Z)

print(X[:,1])
```

## Matplotlib

```python
# import numpy as np (already imported above,
# but you need numpy in the following steps)
import matplotlib.pyplot as plt
```

超基本的な使い方は、`x`と`y`を設定して`plot`に突っ込むだけ。  
突っ込む先が変われば表現する方法も変わる（ヒストグラムとか）

```python
# creating data
x = np.arange(0, 6, 0.1)  #from 0 to 6 and the interbal is 0.1
y = np.sin(x)

# plotting
plt.plot(x, y)
plt.show()
```

複数のグラフを並べるときは以下のように複数`plot`してから`show`すれば良い。
グラフ名の設定もお忘れなく。

```python
# creating data
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# plotting
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos')
plt.xlabel('this is a label of x')
plt.ylabel('you can set your own label here')
plt.title('sin vs cos')
plt.legend()  # this shows how to tell sin from cos (box in the left bleow)
plt.show()
```

`pyplot`の`imshow()`というメソッドで画像を表示したり、`matplotlib.image`というメソッドで画像を読み込んだりできるらしい。
